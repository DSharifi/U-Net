```python
import os
from collections import Counter
import matplotlib.pyplot as plt
from PIL import Image

cache_imgs = []
dir_name = "data/test_x/"
for filename in sorted(os.listdir(dir_name)):
    if filename.endswith(".jpg"): 
        img = Image.open(dir_name + filename)
        size = img.size
        cache_imgs.append(size)


cache_masks = []
dir_name = "data/test_y/"

for filename in sorted(os.listdir(dir_name)):
    if filename.endswith(".png"): 
        img = Image.open(dir_name + filename)
        size = img.size
        cache_masks.append(size)

plt.plot(cache_imgs)
plt.show()
            

#print("Unique Image sizes: " + str(sorted(cache_imgs)))
print(Counter(cache_imgs).values())
```
